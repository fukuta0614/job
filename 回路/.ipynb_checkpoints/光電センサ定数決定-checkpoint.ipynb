```python
import numpy as np
```

```python
def symbol2capacitance(symbol):
    return int(symbol[:2]) * 10 ** (int(symbol[2]) - 12)

def calc_amp_rate(r1, c1, r2, c2, f=2000):
    w = 2 * np.pi * f
    c1 = symbol2capacitance(c1)
    c2 = symbol2capacitance(c2)
    return w * r2 * c1 / np.sqrt((1 + (w*r2*c2)**2)*(1 + (w*r1*c1)**2))
    
def cutoff_freq(r, c):
    c = symbol2capacitance(c)
    return 1 / (2*np.pi*r*c)
```

```python
C1 = '104'
R1 = 1000
C2 = '102'
R2 = 510000
```

```python
def func(r1, c1, r2, c2):
    x = calc_amp_rate(r1, c1, r2, c2)
    y = cutoff_freq(r1, c1)
    z = cutoff_freq(r2, c2)
    return (x, y, z), (x-100)**2 * 20**2 + (y-1000)**2 + (z-3000)**2
    
```

```python
capaciters = ['010', '050', '101', '471', '102', '472', '103', '104', '105']
registors1 = [560, 680, 820, 1000, 1200, 1500, 2200, 2700, 3300, 3900, 4700, 5600, 6800, 8200]
registors2 = list(range(200000, 800000, 20000))
```

```python
res = []
for r1 in registors1:
    for r2 in registors2:
        for c1 in capaciters:
            for c2 in capaciters:
                (x,y,z), score = func(r1, c1, r2, c2)
                res.append(((r1, c1, r2, c2), {'増幅率':x, 'カットオフ1':y, 'カットオフ2':z}, score))
```
